```python
import pandas as pd
```

```python
course_dataframe = pd.read_json("../raw_course_data.json")

exploded_dataframe = course_dataframe.explode('courses')

exploded_dataframe["courseName"] = exploded_dataframe.courses.apply(
    lambda x: x["name"]) 

exploded_dataframe["skills"] = exploded_dataframe.courses.apply(
    lambda x: x["skills"]) 

exploded_dataframe["description"] = exploded_dataframe.courses.apply(
    lambda x: x["description"]) 

exploded_dataframe = exploded_dataframe.drop("courses", axis=1)

exploded_dataframe.head()
```

```python
course_dataframe = pd.read_json("../raw_course_data.json")

df = course_dataframe.explode('courses').reset_index()

normalised_df = pd.json_normalize(df.courses)

concat_normalised_df = pd.concat([df, normalised_df], axis=1)

concat_normalised_df.drop('courses', axis=1)

exploded_locations = concat_normalised_df.explode('locations').reset_index().drop(['level_0', 'index', 'courses'], axis=1).explode('skills').reset_index()

final = exploded_locations.drop(['index'], axis=1)

final
```
